### 1. Что такое Linux?

→ Прежде чем приступать к знакомству с нейронными сетями, давайте отойдём на пару шагов назад и посмотрим на своё рабочее пространство. С какой операционной системой вы работаете? Скорее всего, это Windows или Mac OS. Однако, как будущим дата сайентистам, вам просто необходимо подружиться с операционной системой Linux, чему и будет посвящён данный обучающий модуль.
В случае, если у вас установлен Linux, и вы знакомы с этой системой, мы советуем вам всё же ознакомиться с материалами модуля, чтобы закрепить знания и ничего не упустить.

Linux — это open source-операционная система, относящаяся к семейству Unix-систем (вместе с Mac OS и Android). 

Linux ОС была разработана в 1991 году финно-американским программистом Линусом Торвальдсом и с тех пор только набирает популярность. На сегодняшний день подавляющее большинство серьёзных вычислительных платформ и серверов работает именно на Linux. Именно поэтому любому уважающему себя IT-специалисту рекомендуется освоить Linux.

Поскольку Linux — это open source-операционная система, она существует в огромном количестве версий. Эти версии называются дистрибутивами. Самыми популярными из них являются Ubuntu, Manjaro и Mint.  В нашем курсе мы будем работать с дистрибутивом Ubuntu, однако вы можете выбрать тот вариант системы, который больше всего отвечает вашим требованиям. 

→ Подробнее об отличиях можно прочитать здесь.

#### ПОЧЕМУ LINUX ПОПУЛЯРЕН?

Подавляющее большинство пользователей Linux — разработчики, инженеры, учёные и другого рода профессионалы. Они выбирают эту систему, поскольку:

под Linux написаны тысячи специализированных программ;
при правильной настройке она быстрее, стабильнее и безопаснее, чем Windows и Mac OS;
в ней предусмотрен удобный механизм обновления — можно обновить только необходимые компоненты;
можно полностью настроить систему под свои вкусы.

#### ЗАЧЕМ LINUX СПЕЦИАЛИСТУ В DATA SCIENCE?

Большинство инструментов разработки уже встроено в систему.
Почти все языки программирования, библиотеки и фреймворки разработаны под Linux.
Для любой деятельности можно выбрать удобный набор инструментов.
Большинство операций в Unix-системах выполняются не с помощью графического интерфейса (GUI), к которому мы так привыкли в Windows, а с помощью командной строки (CLI), которая в Unix-системах именуется терминалом.

#### ЗАЧЕМ НУЖНА КОМАНДНАЯ СТРОКА?

Использование терминала вместо графического интерфейса начинающему пользователю Linux может показаться излишне изощренным способом взаимодействия с ОС. Но у этого подхода существует ряд неоспоримых преимуществ:

Скорость взаимодействия пользователя с ОС выше из-за отсутствия подтормаживания графического интерфейса.
Скорость взаимодействия увеличивается по мере освоения основных команд терминала.
Большинство команд терминала позволяют «лёгким движением руки» произвести сложную операцию над файлами и папками, например найти и удалить в папке все текстовые файлы, в которых присутствует слово «привет».
Использование командной строки для удалённого администрирования серверов сокращает интернет-трафик, поэтому администрировать сервера можно даже с мобильного телефона в сети 3G.
На серверах обычно нет графического интерфейса (GUI), поэтому взаимодействие с ними осуществляется через терминал.
Командная строка (CLI) позволяет автоматизировать повторяющиеся действия.
Как видите, командная строка операционной системы Linux является эффективным и востребованным инструментом для IT-специалистов, поэтому в предстоящем модуле вам предстоит познакомиться с ней поближе.

### 3. Знакомство с Linux


→ Что ж, у нас есть система, с которой мы можем работать — давайте не будем терять ни секунды! Первым делом, если вы установили систему с графической оболочкой через виртуальную машину или облако, стоит освоиться с интерфейсом. 

Файловая система в Linux несколько отличается от привычной нам в Windows и MacOS. Давайте разберёмся, как она выглядит и из чего состоит.

#### СТРУКТУРА ФАЙЛОВОЙ СИСТЕМЫ LINUX 

Директория в Linux — это список соответствий имени файла и узла, где располагается вся информация о файле. 

**$ ls -l**

Название колонки | Описание
:-- | :--
__/__ | корневой каталог файловой системы. Директории в нём (см. ниже) могут располагаться на различных физических дисках и разделах. В Linux про диски С: , D: и т.д. можно забыть.  
__/boot__ | инструментарий запуска ОС, в повседневной работе не требуется. Ядро системы.  
__/etc__ | директория, где в текстовых конфигурационных файлах хранятся системные настройки, чаще всего в формате "Parameter Value". Также — основные настройки некоторых приложений. Аналог Windows Registry.  
__/root__ | домашний каталог пользователя root.  
__/bin__ | здесь хранятся исполняемые файлы для утилит системы.  
__/lib__ | здесь хранятся системные библиотеки.  
__/usr__ | сюда устанавливаются пользовательские приложения. Аналог Program Files в Windows.  
__/usr/lib__ | здесь обычно хранятся библиотеки установленных приложений.  
__/usr/bin__ —|здесь обычно хранятся исполняемые файлы для запуска приложений.  
__/dev__ | в *nix используется метафора «всё есть файл». Поэтому каждому устройству ставится в соответствие псевдофайл, который служит для обмена данными с операционной системой и складывается в эту директорию.
__/proc__ | аналогичная схема используется и для процессов системы. Им также соответствуют псевдофайлы в этом каталоге, как это делается для устройств.  
__/tmp__ | директория временных файлов. Аналог Windows\Temp.  
__/home__ | директория, где хранится пользовательская информация. Для каждого пользователя, заведённого в систему, создаётся своя поддиректория по его login, например /home/myuser1. Это аналог Users\myuser1 в Windows. В этих каталогах хранятся файлы пользователя и его индивидуальные настройки.

### 4. Команды и типы команд

Что ж, на структуру файловой системы посмотрели — пора переходить к знакомству с командами!

→ Для взаимодействия с системой Linux используются так называемые bash-команды. 

**Bash** — это оболочка командной строки (командный интерпретатор) в системах Linux, которая поддерживает bash-команды и bash-скрипты (набор bash-команд).

#### КОНВЕНЦИИ КОМАНД LINUX

Для начала можно познакомиться с базовыми принципами работы с командной строкой — так называемыми **конвенциями**. Это поможет вам быстрее разобраться с командами.

1. Первое слово команды — это сама команда (что мы просим сделать)
2. За командой следуют опции, или параметры. Они начинаются со знака минус (-). Регистр имеет значение!
3. Параметры обозначаются одной буквой. Регистр имеет значение!
4. Параметры могут объединяться: -la = -l -a
5. Параметры могут следовать в любом порядке.
6. Списки разделяются пробельными символами.
7. man — подробная справка по любой команде.  

Пример простой команды, которая показывает, в какой директории мы сейчас находимся: **pwd** (present working directory)

_bin>_ pwd

 Воспользуемся командой **ls**, которая показывает список всех файлов и папок в текущей директории.

_bin>_ ls

В этом случае мы использовали только название команды для её вызова, но мы можем не ограничиваться этим и уточнить работу команды, используя **встроенные ключи** (как команда должна быть применена) и **аргументы** (к чему команда должна быть применена).

Предположим, мы хотим вывести содержимое директории в виде списка и заодно посмотреть дополнительную информацию (дату создания, размер и другое). В таком случае мы добавим к команде ls ключ -l (от англ. list). 



_bin>_ ls -l

Если нам требуется вывести в том числе и скрытые файлы и папки, то мы можем добавить ещё один ключ -a (от англ. all): 



ls -l -a  
ls -la

Ключи имеют короткую и/или длинную форму. Ключ команды ls -l имеет только короткую форму, а ключ -a имеет как короткую, так и длинную формы.  
Порядок ключей также может быть любым. Такие команды будут равнозначны: 





ls -l --all  
ls -l -a  
ls -la  
ls-al

**Аргументом** для команды ls будет служить название директории. Например, чтобы посмотреть содержимое папки Documents, мы напишем команду:

ls -l Documents

Чтобы посмотреть все возможности команды ls, давайте вызовем справку с помощью команды *man* (от англ. manual):



_bin>_ man ls

Чтобы выйти из справки, нажмите q.

#### ТИПЫ КОМАНД

→ Все команды в Linux можно поделить на два типа — базовые и те, которые управляют заранее установленными программами.

1. Базовые команды присутствуют в большинстве дистрибутивов Linux-систем: ls, pwd, mkdir, rm и так далее.
2. Команды, которые управляют заранее установленными программами: python, docker, hadoop и так далее.  

**Базовые команды** — это такие команды, которые являются составляющей частью командной оболочки, доступны из любой папки практически в любой системе Linux. Ниже представлен неполный перечень таких команд.

Команда | Описание
:-- | :--
__cd data__ |	Позволяет перейти в директорию data
__cd ..__ |	Перейти на директорию выше
__cd ~__ |	Перейти в домашнюю директорию: /home/username/
__cd -__ |	Перейти в предыдущую папку
__pwd__ |	Показать текущую директорию
__mkdir data__ |	Создать каталог с именем data
__touch text.txt__ |	Создать файл с именем text.txt
__rm -r data__ |	Удалить каталог с именем data
__rm text.txt__ |	Удалить файл с именем text.txt

#### КАКАЯ ИНФОРМАЦИЯ НАХОДИТСЯ В ЭТИХ СТРОКАХ

**Строка 1**

- текущее время (top обновляет экран раз в 3 секунды);
- так называемый uptime — сколько времени система работает с последней перезагрузки (5 дней и 14 часов);
- средняя загруженность системы за последние 1, 5 и 15 минут.  

**Строка 2**

- количество потоков в разбивке: работающие, спящие, остановленные и зомби. 

**Строка 3**

- использование CPU в %% за последний интервал в разбивке по циклам; важные циклы:
- us — работа в пользовательском режиме;
- sy — работа в режиме ядра;
- id — «ничегонеделание»;
- wa — ожидание окончания ввода-вывода.  

- здесь показывается сводная информация по всем процессорам системы.  

**Строка 4**

- информация об оперативной памяти (в данном случае в килобайтах)
- всего в системе;
- свободно;
- используется;
- используется кэшем системы.  

**Строка 5**

- информация о swap (аналогично);
- последнее поле avail — память, доступная новым процессам.  

**Строка 6**

- заголовок колонок.  

**Строки 7 и далее**

- информация о процессах, выполнявшихся в последнем отрезке времени. По умолчанию выводятся следующие поля:
- PID — идентификатор процесса
- USER — владелец процесса (effective user ID)
- PR — приоритет процесса
- NI — изменение приоритета (отрицательное значение — был увеличен, положительное — уменьшен,  — не менялся)
- VIRT — общий объём виртуальной памяти, которую занимает процесс (килобайты)
- RES — размер резидентной (не отсвоппированной) области памяти процесса (килобайты)
- SHR — объём разделяемой памяти процесса
- S — статус процесса (S — sleeping, R — running. Другие менее интересны)
- %CPU — процент использования CPU (как правило, одного) процессом
- %MEM — текущая доля доступной физической памяти, занимаемая процессом
- TIME+ — общее время работы с начала запуска процесса (с сотыми долями секунды)
- COMMAND — выполняемая команда  

#### КОММЕНТАРИИ

На что стоит обращать внимание в первую очередь при работе с top:

- id в строке 3 — это процент времени, когда система была «свободна».
- wa в строке 3 — сколько-нибудь существенные значения здесь говорят о том, что в системе происходит какой-то интенсивный ввод-вывод. Это, скорее всего, отражается на быстродействии системы (значения больше  приводят к существенным замедлениям, ощущаемым даже в командной строке).
- avail в строке 5 — доступная память. Если её мало, скорее всего будет swap.
- верхние строки в табличной части (с информацией о процессах) — постоянно присутствующие там процессы, скорее всего, являются причиной «торможения» системы (если таковое имеет место). 

Отметим, что 100%-ное использование процессом процессора (по информации в табличной части) — это неплохой результат, если это происходит с вашим приложением. Это означает, что оно эффективно работает. Для приложений, которые работают с данными, такой результат практически недостижим: ввод-вывод существенно понижает %% использования CPU.

Также заметим, что %% использования CPU в табличной части показывает загруженность одного процессора, %% использования CPU в «шапке» — общую картину (всех процессоров).

- Часть проблем в Linux возникает из-за некорректной персонализации.

#### UID И LOGIN

Пользователь в Linux — это уникальная комбинация **UID (уникальный идентификатор, число)** и **login (строка)**. Отображается login, а системой используется UID. В кластере и логины, и UID-пользователей должны быть одинаковыми. 

Пользователи объединяются в группы, при этом пользователь может входить в одну или несколько групп. У группы также есть уникальная комбинация **GID (числа)** и **имени (строки)**. Группы активно используются при работе с файлами. 

Информация о UID и login пользователя, а также принадлежности к его первичной группе, хранится в файле __/etc/passwd__.

ID и имя группы, а также список пользователей, которые входят в эту группу, хранятся в файле __/etc/group.__

При входе пользователя в систему происходит поиск логина и назначение соответствующего ему ID. ID текущего пользователя можно сменить, если есть на это полномочия

- перечень групп, в которые входит данный пользователь (идентификаторы и имена). В данном случае мы имеем 7 групп, включая первичную.

#### ФАЙЛ /ETC/PASSWD

Информация о пользователях хранится в файле /etc/passwd.

head -5 /etc/passwd

root:x:0:0:root:/root:/bin/bash
daemon:x:1:1:daemon:/usr/sbin:/usr/sbin/nologin
bin:x:2:2:bin:/bin:/usr/sbin/nologin
sys:x:3:3:sys:/dev:/usr/sbin/nologin
sync:x:4:65534:sync:/bin:/bin/sync

В каждой строке файла содержится информация об одном пользователе, поля в строке разделены двоеточиями.

Что мы видим (на примере первой строки):

первое поле — логин пользователя (root);
второе поле — пароль (изначально он хранился прямо здесь в зашифрованном виде, потом его перенесли в другой файл);
третье поле — uid (0);
четвертое поле — gid (0);
пятое поле — комментарий (обычно сюда пишут полное имя пользователя);
шестое поле — домашняя директория (/root). Та директория, которая будет текущей после логина пользователя;
седьмое поле — команда, которая будет выполнена после успешного логина пользователя (/bin/bash).
Обратите внимание, как помечаются пользователи, которые не могут «войти в систему» (то есть работать интерактивно). Команда, которая выполняется после их логина — /usr/sbin/nologin. Подробнее поговорим об этом ниже.

#### ПОЛЬЗОВАТЕЛЬ ROOT

Пользователь root (uid 0) — специальный суперпользователь. Он существует всегда, от его имени работает большинство системных процессов, он является владельцем большинства системных файлов.

Права пользователя root не проверяются.

### 8. Файлы, директории и права доступа

→ Теперь, когда мы разобрались с пользователями и группами в системе Linux, пора познакомиться с командами, которые касаются прав доступа к файлам и директориям.

В Linux права доступа делятся на три категории.

- владелец (user);
- группа (group);
- все остальные (others).  

Виды доступа (что мы можем разрешить делать):

- чтение (r);
- запись (w);
- исполнение (x). Для директорий право на исполнение означает право доступа к файлам директории.  

**Право на чтение директории** — это право на чтение/запись/изменение этого каталога. Когда мы даём право на выполнение директории, мы эмулируем возможность или невозможность доступа к файлу, который находится не внутри каталога, а внутри файловой системы.

**Добавить нового пользователя:**

```python
sudo adduser thatguy (имя пользователя)
```

**Передать права:**

```python
sudo chown thatguy 123.ipybn
```

```python
sudo addgroup somegroup
```

**Добавление пользователя в группу:**

```python
sudo usermod -G me (название группы) thatguy
```

#### КОМАНДА CHMOD

**Назначение: изменение прав доступа к файлу.**

В качестве параметров задаются права (точнее то, как их нужно изменить) и имя файла (список имён файлов).

Важно: не работает для файлов, которыми владеет Windows (именно поэтому в примерах ниже используется /tmp/123.txt)

Для описания изменений в правах доступа можно использовать две разных нотации:

символическую (вида u+rw);
цифровую (вида 0666).
Рассмотрим каждую из них.

#### СИМВОЛИЧЕСКАЯ НОТАЦИЯ

Необходимо указать:

- для какой категории прав доступа (владелец, группа или все остальные) происходит изменение прав;
- какого рода изменение требуется (добавить права, отнять права или установить права);
- какой вид доступа участвует в изменениях.
- Категория прав задается одной буквой:

Буква | Категория права
:-- | :--
__u__ |	изменение прав для владельца
__g__ |	изменение прав для группы
__o__ |	изменение прав для всех остальных
__a__ |изменение прав для всех категорий (пользователь, группа и все остальные)

Характер изменения — это один из символов:



Символ | Характер изменения
:-- | :--
__+__ |	добавить права
__-__ |	отнять права
__=__ |	установить права


Вид доступа — это одна буква:

Буква | Вид доступа
:-- | :--
__r__ |	права на чтение
__w__ |	права на запись
__x__ |	права на исполнение (для директории — чтение каталога)


Изменять права доступа к файлу может только владелец файла (или суперпользователь). Не будет ошибкой дважды отнять (отнять право, которого и так не было) или дважды дать (дать право, которое и так есть) право доступа.

```python
chmod a-w /tmp/123.txt
```

В данном случае мы «отняли» (-) у всех (a) права на запись (w) в файл 123.txt, находящийся в директории /tmp.



```python
chmod u+w /tmp/123.txt
```

В данном случае мы «добавили» (+) владельцу файла (u) право на запись (w) в файл 123.txt, находящийся в директории /tmp.

```python
chmod g=rx /tmp/123.txt
```

В данном случае мы «установили» (=) группе владельца файла (g) право на чтение и выполнение (rx) файла 123.txt, находящегося в директории /tmp.

#### ЦИФРОВАЯ НОТАЦИЯ

Права доступа можно задавать в цифровой нотации. Такая запись может быть удобнее символической, так как является более лаконичной. Для цифровой нотации используются следующие значения прав доступа:

Цифра | Значение
:-- | :--
__1__ | исполнение
__2__ | запись
__3__ | чтение

Сумма этих чисел представляет из себя устанавливаемое право доступа. Например:
    
3 = 1 + 2 — право на запись и исполнение.  
6 = 2 + 4 — право на чтение и запись.

Для каждой категории доступа (владелец, группа и остальные) необходимо указать одну цифру (от 0 до 7). В случае если вы указываете меньшее количество цифр, система будет читать эту запись так, как если бы все неуказанные цифры были нулями. При этом условные нули ставятся перед введённой цифрой. Например:

666 — это права на чтение и запись для владельца, группы и остальных.  

4 — это право на чтение для остальных, владелец и группа не имеют прав на файл (как если бы вы ввели 004).  

400 — это право на чтение для владельца, группа и остальные не имеют прав на файл.  

С помощью цифровой нотации можно только установить права доступа (то есть нельзя добавить или отнять).

```python
chmod 666 /tmp/123.txt
```

#### РЕКУРСИВНОЕ ИЗМЕНЕНИЕ ПРАВ -R

С помощью этого параметра команды происходит рекурсивное изменение прав (то есть изменение прав в текущей директории во всех вложенных директориях).

В этом случае имя файла должно задавать маску (содержать специальные символы * или ?) или быть директорией (в этом случае будут изменены права для всех файлов в директории).

```python
chmod a+r /tmp/ -R
```

#### КОМАНДЫ CHOWN И CHGRP

Назначение: изменить владельца или группу владельца файлов, заданных в качестве параметров.

Операцию может выполнить тот, кто является владельцем файла (или суперпользователь).

Важно! После смены владельца текущий пользователь может потерять право дальнейших манипуляций файлом (обратно «вернуть» файл себе уже не получится).

```python
chown daemon /tmp/124.txt
chown mk /tmp/124.txt
```

chown: changing ownership of '/tmp/124.txt': Operation not permitted  

Можно изменить владельца и группу одновременно. Для этого необходимо указать их через "." в качестве параметра команды chown.

```python
chown mk.mk /tmp/123.txt
```
