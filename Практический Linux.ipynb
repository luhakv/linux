### 1. Что такое Linux?

→ Прежде чем приступать к знакомству с нейронными сетями, давайте отойдём на пару шагов назад и посмотрим на своё рабочее пространство. С какой операционной системой вы работаете? Скорее всего, это Windows или Mac OS. Однако, как будущим дата сайентистам, вам просто необходимо подружиться с операционной системой Linux, чему и будет посвящён данный обучающий модуль.
В случае, если у вас установлен Linux, и вы знакомы с этой системой, мы советуем вам всё же ознакомиться с материалами модуля, чтобы закрепить знания и ничего не упустить.

Linux — это open source-операционная система, относящаяся к семейству Unix-систем (вместе с Mac OS и Android). 

Linux ОС была разработана в 1991 году финно-американским программистом Линусом Торвальдсом и с тех пор только набирает популярность. На сегодняшний день подавляющее большинство серьёзных вычислительных платформ и серверов работает именно на Linux. Именно поэтому любому уважающему себя IT-специалисту рекомендуется освоить Linux.

Поскольку Linux — это open source-операционная система, она существует в огромном количестве версий. Эти версии называются дистрибутивами. Самыми популярными из них являются Ubuntu, Manjaro и Mint.  В нашем курсе мы будем работать с дистрибутивом Ubuntu, однако вы можете выбрать тот вариант системы, который больше всего отвечает вашим требованиям.

#### ПОЧЕМУ LINUX ПОПУЛЯРЕН?

Подавляющее большинство пользователей Linux — разработчики, инженеры, учёные и другого рода профессионалы. Они выбирают эту систему, поскольку:

под Linux написаны тысячи специализированных программ;
при правильной настройке она быстрее, стабильнее и безопаснее, чем Windows и Mac OS;
в ней предусмотрен удобный механизм обновления — можно обновить только необходимые компоненты;
можно полностью настроить систему под свои вкусы.

#### ЗАЧЕМ LINUX СПЕЦИАЛИСТУ В DATA SCIENCE?

Большинство инструментов разработки уже встроено в систему.
Почти все языки программирования, библиотеки и фреймворки разработаны под Linux.
Для любой деятельности можно выбрать удобный набор инструментов.
Большинство операций в Unix-системах выполняются не с помощью графического интерфейса (GUI), к которому мы так привыкли в Windows, а с помощью командной строки (CLI), которая в Unix-системах именуется терминалом.

#### ЗАЧЕМ НУЖНА КОМАНДНАЯ СТРОКА?

Использование терминала вместо графического интерфейса начинающему пользователю Linux может показаться излишне изощренным способом взаимодействия с ОС. Но у этого подхода существует ряд неоспоримых преимуществ:

Скорость взаимодействия пользователя с ОС выше из-за отсутствия подтормаживания графического интерфейса.
Скорость взаимодействия увеличивается по мере освоения основных команд терминала.
Большинство команд терминала позволяют «лёгким движением руки» произвести сложную операцию над файлами и папками, например найти и удалить в папке все текстовые файлы, в которых присутствует слово «привет».
Использование командной строки для удалённого администрирования серверов сокращает интернет-трафик, поэтому администрировать сервера можно даже с мобильного телефона в сети 3G.
На серверах обычно нет графического интерфейса (GUI), поэтому взаимодействие с ними осуществляется через терминал.
Командная строка (CLI) позволяет автоматизировать повторяющиеся действия.
Как видите, командная строка операционной системы Linux является эффективным и востребованным инструментом для IT-специалистов, поэтому в предстоящем модуле вам предстоит познакомиться с ней поближе.

### 3. Знакомство с Linux


→ Что ж, у нас есть система, с которой мы можем работать — давайте не будем терять ни секунды! Первым делом, если вы установили систему с графической оболочкой через виртуальную машину или облако, стоит освоиться с интерфейсом. 

Файловая система в Linux несколько отличается от привычной нам в Windows и MacOS. Давайте разберёмся, как она выглядит и из чего состоит.

#### СТРУКТУРА ФАЙЛОВОЙ СИСТЕМЫ LINUX 

Директория в Linux — это список соответствий имени файла и узла, где располагается вся информация о файле. 

**$ ls -l**

Название колонки | Описание
:-- | :--
__/__ | корневой каталог файловой системы. Директории в нём (см. ниже) могут располагаться на различных физических дисках и разделах. В Linux про диски С: , D: и т.д. можно забыть.  
__/boot__ | инструментарий запуска ОС, в повседневной работе не требуется. Ядро системы.  
__/etc__ | директория, где в текстовых конфигурационных файлах хранятся системные настройки, чаще всего в формате "Parameter Value". Также — основные настройки некоторых приложений. Аналог Windows Registry.  
__/root__ | домашний каталог пользователя root.  
__/bin__ | здесь хранятся исполняемые файлы для утилит системы.  
__/lib__ | здесь хранятся системные библиотеки.  
__/usr__ | сюда устанавливаются пользовательские приложения. Аналог Program Files в Windows.  
__/usr/lib__ | здесь обычно хранятся библиотеки установленных приложений.  
__/usr/bin__ —|здесь обычно хранятся исполняемые файлы для запуска приложений.  
__/dev__ | в *nix используется метафора «всё есть файл». Поэтому каждому устройству ставится в соответствие псевдофайл, который служит для обмена данными с операционной системой и складывается в эту директорию.
__/proc__ | аналогичная схема используется и для процессов системы. Им также соответствуют псевдофайлы в этом каталоге, как это делается для устройств.  
__/tmp__ | директория временных файлов. Аналог Windows\Temp.  
__/home__ | директория, где хранится пользовательская информация. Для каждого пользователя, заведённого в систему, создаётся своя поддиректория по его login, например /home/myuser1. Это аналог Users\myuser1 в Windows. В этих каталогах хранятся файлы пользователя и его индивидуальные настройки.
